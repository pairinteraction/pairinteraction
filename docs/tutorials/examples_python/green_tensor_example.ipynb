```python
from pairinteraction.green_tensor import (
    GreenTensorCavity,
    GreenTensorFreeSpace,
    GreenTensorSurface,
)
```

```python
gt = GreenTensorFreeSpace()
gt.set_atom_positions([0, 0, 0], [5, 0, 0], "micrometer")
gt.set_electric_permitivity(1)

tensor = gt.get_dipole_dipole(omega=1, unit="GHz")
```

```python
gt = GreenTensorSurface()
gt.set_atom_positions([0, 0, 0], [5, 0, 0], "micrometer")
gt.set_electric_permitivity(1)
gt.set_surface_position(z=1, unit="centimeter")
gt.set_electric_permitivity_surface(3 + 2j)

tensor = gt.get_dipole_dipole(omega=1, unit="GHz")
```

```python
gt = GreenTensorCavity()
gt.set_atom_positions([0, 0, 0], [5, 0, 0], "micrometer")
gt.set_electric_permitivity(1)
gt.set_surface_positions(z1=-1, z2=1, unit="centimeter")
gt.set_electric_permitivity_surfaces(epsilon1=3 + 2j, epsilon2=4 + 1j)

tensor = gt.get_dipole_dipole(omega=1, unit="GHz")
```

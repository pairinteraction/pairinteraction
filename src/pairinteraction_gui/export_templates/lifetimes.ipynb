# Lifetime Calculation

```python
# Uncomment the next line if you have ipympl installed and want interactive plots
# %matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

import pairinteraction as pi
```

## State of Interest

```python
ket = pi.KetAtom($SPECIES_0$QUANTUM_NUMBERS_0)
temperature = $TEMPERATURE

print(f"State of interest: {ket}")
print(f"Temperature: {temperature} K")
```

## Lifetime and Transition Rates

```python
kets_sp, rates_sp = ket.get_spontaneous_transition_rates(unit="1/ms")
kets_bbr, rates_bbr = ket.get_black_body_transition_rates(temperature, "K", unit="1/ms")
lifetime = ket.get_lifetime(temperature, temperature_unit="K", unit="mus")

print(f"Lifetime: {lifetime:.3f} mus")
print(f"Number of spontaneous transitions: {len(kets_sp)}")
print(f"Number of BBR transitions: {len(kets_bbr)}")
```

```python
n_list = np.arange(0, np.max([state.n for state in list(kets_sp) + list(kets_bbr)] + [0]) + 1)
sorted_rates = {}

for key, kets, rates in [("SP", kets_sp, rates_sp), ("BBR", kets_bbr, rates_bbr)]:
    sorted_rates[key] = {n: [] for n in n_list}
    for state, rate in zip(kets, rates, strict=True):
        sorted_rates[key][int(state.n)].append((str(state), float(rate)))

rates_summed = {
    key: np.array([sum(rate for _, rate in sorted_rates[key][n]) for n in n_list]) for key in sorted_rates
}
```

## Plot

```python
fig, ax = plt.subplots()

ax.bar(n_list, rates_summed["SP"], label="Spontaneous Decay", color="blue", alpha=0.8)
ax.bar(n_list, rates_summed["BBR"], label="Black Body Radiation", color="red", alpha=0.8)

ax.set_xlabel("Principal Quantum Number $n$")
ax.set_ylabel(r"Transition Rates (1 / ms)")
ax.legend()

plt.show()
```
